## Encontrar Qt-desginer

<p style="text-align: justify;">
En primer lugar, deben asegurarse de tener instalado Qtdesigner, para esto deben escribir lo 
siguiente en consola:
</p>

`pip install PyQt5-tools`

<p style="text-align: justify;">
El siguiente paso es enontrar Qtdesginer, ya que no aparecera por defecto en el escritorio ni en un lugar de facil acceso, pueden buscar escribir designer en el buscador para encontrarlo como se muestra en la siguiente imagen:
</p>



<p style="text-align: center;">
Fuente: Comentario de [@stephichau](https://github.com/stephichau) en issue [#516](https://github.com/IIC2233/Syllabus/issues/516) del semestre 2017-1
</p>


En caso de que no lo encuentren, pueden hacerlo en la siguiente ruta (el programa se llama designer.exe):

**C:\Users\[Ac va su nombre de usuario del Pc]\AppData\Local\Programs\Python\Python35\Lib\site-packages\pyqt5-tools\designer**

Una vez que encuentren el programa se recomienda que dejen un acceso directo en el escritorio para que sea de rpido acceso.
