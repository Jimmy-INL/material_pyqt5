## Abriendo y conociendo Qt-desginer

La primera ventana con la que se encontrarn les preguntara que tipo de ventana quieren comenzar a crear, como muestra la siguiente imagen:



Luego se ecnontraran con la principal ventana de trabajo de desginer, en la que tendrn distintas barras de herramientas como se muestra a continuacin:



### 1-. Ventana de edicin

<p style="text-align: justify;">

Es el lugar en donde se edita el widget o ventana que estas creando, en esta zona debes agregar todos los widgets que te ofrece PyQt5 para formar la ventana tal cual quieres.

</p>

### 2-. Caja de widgets

<p style="text-align: justify;">

En esta barra de herramientas encuentras los widgets que puedes agregar a la ventana que ests creando, basta simplemente con pinchar uno de ellos y arrastrarlo hasta donde quieras que quede en tu ventana. Puede ser de mucha utilidad que explores los diferentes widgets que existen ac, ya que pueden facilitarte el desarrollo de tus tareas al exisitir muchos que cumplen una funcin especifica.

</p>

### 3-. Inspector de Objetos

<p style="text-align: justify;">

En esta barra de herramientas puedes ver todos los widgets que se encuentran actualmente en tu ventana. Adems, puedes observar que widget se encuentra al interior de otro widget y puedes seleccionar cualquiera de ellos para examinar sus propiedades en el editor de propiedades.

</p>

### 4-. Editor de Propiedades


<p style="text-align: justify;">

En esta barra de herramientas puedes encontrar todas las propiedades del objeto que tengas seleccionado, ademas encontrars separadas las propiedades del objeto que son heredadas de las nicas de si mismo. Todos los cambios que realices en esta seccin quedaran guardados en el archivo que guardes y que cargues posteriormente en el cdigo, por lo que es una manera sencilla y grfica de controlar que propiedades tienen los widgets que ests utilizando. 

Por ejemplo, si seleccionas un Spin Box, podrs notar que tiene propiedades de QObject (nombre), QWidget, QAbstractSpinBox y QSpinBox, las que si cambias podrs verlas reflejadas inmediatamente en el Spin Box en la ventana misma (cambiando los valores de geometry, por ejemplo).

</p>


### 5-. Editor de Seales/ Editor de Acciones / Navegador de Recursos

<p style="text-align: justify;">

En el editor de seales puedes crear de manera rpida seales e indicar que tipo de objeto las emitir y que tipo de objeto las recibira.

</p>

### 6-. Herramientas del archivo y Modo de edicion

<p style="text-align: justify;">

Las primeras herramientas son las usuales de todos los programas para poder abrir un nuevo archivo, guardar el actual o abrir uno existente. Las otras herramientas son para modificar lo que se est editando en la ventana de edicin, por defecto est seleccionado el modo de edicin de widgets pero si presionamos los otros botones, podrn notar que pueden modificar las seales de los objetos 

</p>
