### 1-. Ventana de edición

<p style="text-align: justify;">

Es el lugar en donde se edita el widget o ventana que estas creando, en esta zona debes agregar todos los widgets que te ofrece PyQt5 para formar la ventana tal cual quieres.

</p>

### 2-. Caja de widgets

<p style="text-align: justify;">

En esta barra de herramientas encuentras los widgets que puedes agregar a la ventana que estás creando, basta simplemente con pinchar uno de ellos y arrastrarlo hasta donde quieras que quede en tu ventana. Puede ser de mucha utilidad que explores los diferentes widgets que existen acá, ya que pueden facilitarte el desarrollo de tus tareas al exisitir muchos que cumplen una función especifica.

</p>

### 3-. Inspector de Objetos

<p style="text-align: justify;">

En esta barra de herramientas puedes ver todos los widgets que se encuentran actualmente en tu ventana. Además, puedes observar que widget se encuentra al interior de otro widget y puedes seleccionar cualquiera de ellos para examinar sus propiedades en el editor de propiedades.

</p>

### 4-. Editor de Propiedades


<p style="text-align: justify;">

En esta barra de herramientas puedes encontrar todas las propiedades del objeto que tengas seleccionado, ademas encontrarás separadas las propiedades del objeto que son heredadas de las únicas de si mismo. Todos los cambios que realices en esta sección quedaran guardados en el archivo que guardes y que cargues posteriormente en el código, por lo que es una manera sencilla y gráfica de controlar que propiedades tienen los widgets que estás utilizando. 

Por ejemplo, si seleccionas un Spin Box, podrás notar que tiene propiedades de QObject (nombre), QWidget, QAbstractSpinBox y QSpinBox, las que si cambias podrás verlas reflejadas inmediatamente en el Spin Box en la ventana misma (cambiando los valores de geometry, por ejemplo).

</p>


### 5-. Editor de Señales/ Editor de Acciones / Navegador de Recursos

<p style="text-align: justify;">

En el editor de señales puedes crear de manera rápida señales e indicar que tipo de objeto las emitirá y que tipo de objeto las recibira.

</p>

### 6-. Herramientas del archivo y Modo de edicion

<p style="text-align: justify;">

Las primeras herramientas son las usuales de todos los programas para poder abrir un nuevo archivo, guardar el actual o abrir uno existente. Las otras herramientas son para modificar lo que se está editando en la ventana de edición, por defecto está seleccionado el modo de edición de widgets pero si presionamos los otros botones, podrán notar que pueden modificar las señales de los objetos de la ventana.

</p>

## Utilizando Qt-Designer

<p style="text-align: justify;">

A continuación crearemos un archivo .ui muy simple que luego pueden exportar a su código para utilizarlo. Queremos crear un formulario de Inicio de Sesión que nos pida usuario y contraseña.

</p>

### Paso 1: Crear archivo .ui en Qt-Designer

<p style="text-align: justify;">

El primer caso consiste en crear el archivo .qt en la aplicación Qt-Designer. En la aplicación estareos creando la interfaz propiamente tal, por lo que debemos agregar los widgets necesarios para crear el formulario. 

Intentaremos crear un formulario similar al siguiente:
</p>

<img src="https://www.solvetic.com/uploads/monthly_11_2014/tutorials-2308-0-47141900-1416871693.jpg" width=350 height=600 alt="Ejemplo Formulario">

<p style="text-align: center;">
Fuente: Imagen de [solvetic](https://www.solvetic.com/uploads/monthly_11_2014/tutorials-2308-0-47141900-1416871693.jpg)
</p>


Para crearlo utilizaremos:

- 1 Vertical Layout que se usará para que los widgets en que se podrá ingresar la contraseña y el usuario queden ordenados uno sobre el otro del mismo tamaño.
- 2 Horizontal Layout uno para el ingreso de la contraseña, en el que se le agregará un Label que indique que en ese espacio se escribe la contraseña y un Line Edit para que el usuario pueda escribirla. El otro se usará de manera analoga para el caso del usuario.
- 1 Label que estará sobre los dos horizontal Layout que dira "Inicio de Sesión".
- 2 Labels que será uno para cada Horizontal Layout que digan "Usuario" y "Contraseña".
- 2 Line Edit para que el usuario pueda escribir su usuario y contraseña respectivamente. 
- 1 Push Button para que se envie la información que el usuario escribio.

Pueden ver la creación del archivo en el siguiente [video](www.google.cl).

### Paso 2: Exportar archivo a un código en python

<p style="text-align: justify;">

Ahora que ya tenemos creada la interfaz en el archivo __formulario.ui__ solo tenemos que importarlo en nuestro código de la siguiente manera :

</p>

```python
from PyQt5.QtWidgets import QApplication
from PyQt5 import uic

# Cargamos la interfaz creada en QtDesigner
qt_name, QtClass = uic.loadUiType("formulario.ui")

# Creamos nuestra ventana del formulario heredando desde la interfaz creada con QTDesigner.
# La función loadUiType retorna una tupla en donde el primer elemento corresponde al
# nombre de la ventana definido en QtDesigner, y el segundo elemento a la clase base de la GUI

# Veamos que es el mismo nombre que definimos en qt-designer
print("nombre de la ventana importada:", qt_name)

class FormWindow(qt_name, QtClass):
    def __init__(self):
        super().__init__()
        self.setupUi(self)
        # Inicializamos la interfaz cargada de qt con el método setupUi


if __name__ == '__main__':
    # Inicializamos la aplicación
    app = QApplication([])
    form = FormWindow()
    form.show()
    app.exec_()
```

### Paso 3: Agregar funcionalidad a Interfaz

Ahora que tenemos creada la interfaz solo nos falta implementar funcionalidad a esta, supongamos que tenemos guardados en un diccionario de los usuarios junto a sus contraseñas. Si el usuario o la contraseña es incorrecta se indicará en el status bar de la ventana que los datos ingresados no son correctos. Si el usuario y la contraseña coinciden se indicara que los datos fueron ingresados correctamente.


```python
from PyQt5.QtWidgets import QApplication
from PyQt5 import uic

# Acá tendremos guardados los datos de usuarios junto a sus contraseñas
# (claramente no es la manera más segura de guardar datos)
users = {"user1": "123456", "user2": "654321", "user3": "123654"}

# Cargamos la interfaz creada en QtDesigner
qt_name, QtClass = uic.loadUiType("formulario.ui")

class FormWindow(qt_name, QtClass):
    def __init__(self):
        super().__init__()
        self.setupUi(self)
        
        # Conectamos el método process_data con el botón que creamos en
        # qt-designer
        self.submit_button.clicked.connect(self.process_data)

    # Definimos el método que será llamado cuando se presione el boton
    def process_data(self):

        # obtenemos el usuario y la contraseña de los Line Edit como pueden
        # obervar tienen el mismo nombre que definimos en el editor de
        # propiedades de qt-designer
        user = self.user_input.text()
        password = self.password_input.text()
        if user in users and password == users[user]:
            self.statusBar().showMessage('Datos ingresados correctamente')
        else:
            self.statusBar().showMessage('Usuario o contraseña incorrectos')

        # borramos los datos que habia ingresado el usuario de los Line Edit
        self.user_input.clear()
        self.password_input.clear()


if __name__ == '__main__':
    app = QApplication([])
    form = FormWindow()
    form.show()
    app.exec_()
```

Como podemos observar tenemos un formulario muy básico que ya es funcional utilizando tanto qt-desginer como código directamente desde python.
